Filtering and Sorting Questions


```python
# 8.	Sort the dataset by Price in descending order.
df.sort_values(by =["Price"],ascending= False)
```

```python
# 9.	Select only the Brand, Model, and Price columns.
df[["Brand","Model","Price"]]
```

```python
# 10.	Find all cars that have been sold.
df[df['Sold']=="Yes"]
```

```python
# 11.	Find all cars that belong to the "BMW" brand.
df[df['Brand']=="BMW"]
```

```python
# 12.	Find all cars that are from the year 2020 or later.
df[df["Year"]>=2020]
```

```python
# 13.	Find all cars that belong to the "North" region and have been sold
df[(df["Region"]=="North") & (df['Sold']=='Yes')]
```

```python
# 14.	Filter cars that have Price greater than $40,000.
df[df["Price"]>40000]
```

Grouping and Aggregation Questions

```python
# 19.	Find the average Price of cars per brand.
df.groupby('Brand')["Price"].mean()
```

```python
# 20.	Count the number of cars in each FuelType category.
df.groupby('FuelType')['FuelType'].count()
```

```python
# 21.	Find the highest Price in each Brand.
df.groupby('Brand')['Price'].max()
```

```python
# 23.	Find the car with the highest ResaleValue.
df['ResaleValue'].max()
df[df['ResaleValue']==df['ResaleValue'].max()]
```

```python
# 24.	Count the number of cars sold and not sold.
df.groupby('Sold')['CarID'].count()
```

```python
# 25.	Find the average Horsepower per Brand.
df.groupby('Brand')['Horsepower'].mean()
```

```python
# 26.	Find the highest ResaleValue per Region.
df.groupby('Region')['ResaleValue'].max()
```

```python
# 27.	Find the lowest Mileage per Brand.
df.groupby('Brand')['Mileage'].min()
```

```python
# 28.	Count the number of cars per Transmission type.
df.groupby('Transmission')['CarID'].count()
```
