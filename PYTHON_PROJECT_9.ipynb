# Define a class called Employee with attributes name, position, and salary.
# Include a method called raise_salary() that takes a percentage increase as an argument and updates the salary accordingly.
# Create an instance of the Employee class and call the raise_salary() method with a 10% increase.

```python
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def raise_salary(self, percentage):
        increase_amount = self.salary * (percentage / 100)
        self.salary += increase_amount
        print(f"Salary increased by {percentage}%. New salary: {self.salary}")

# Creating an instance (object)
emp1 = Employee("Nisha", "Data Analyst", 50000)

# Calling raise_salary() method with 10% increase
emp1.raise_salary(10)
```

# Include methods to deposit money, withdraw money, and check the balance.
# Create an instance of the BankAccount class, deposit $100, withdraw $50, and then print the remaining balance.

```python
class BankAccount:
    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn: ${amount}. Remaining balance: ${self.balance}")

    def check_balance(self):
        print(f"Current Balance: ${self.balance}")

# Creating an instance of BankAccount class
account = BankAccount("12345", "Nisha", 0)

# Performing operations
account.deposit(100)
account.withdraw(50)
account.check_balance()
```

# Define a class named Circle with attributes radius and color.
# Include methods to calculate the area and circumference of the circle.
# Create an instance of the Circle class with radius 5 and color "blue", then print its area and circumference.


```python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color

    def area(self):
        return 3.14159 * (self.radius ** 2)

    def circumference(self):
        return 2 * 3.14159 * self.radius

# Creating an instance of Circle class
circle1 = Circle(5, "blue")

# Printing area and circumference
print("Area:", circle1.area())
print("Circumference:", circle1.circumference())
```

Question 8:
# Create a class named Rectangle with attributes length and width.
# Include methods to calculate the area, perimeter, and diagonal of the rectangle.
# Create an instance of the Rectangle class with length 4 and width 3, then print its area, perimeter, and diagonal.


```python
import math

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def diagonal(self):
        return math.sqrt(self.length**2 + self.width**2)


# Create object of Rectangle class
rect = Rectangle(4, 3)

# Print results
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
print("Diagonal:", rect.diagonal())
```

# Create a class named BankAccount with attributes account_number, account_holder, and balance.
# Include methods to deposit and withdraw money, and to display the account details.
# Create two instances of the BankAccount class, one for a savings account and one for a checking account, and perform deposit and withdrawal operations on each.


```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New Balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn ${amount}. New Balance: ${self.balance}")

    def display_details(self):
        print("\n--- Account Details ---")
        print("Account Number:", self.account_number)
        print("Account Holder:", self.account_holder)
        print("Balance:", self.balance)
        print("------------------------")


# Creating two BankAccount objects
savings = BankAccount("SAV123", "Alice", 500)
checking = BankAccount("CHK456", "Bob", 300)

# Operations on Savings Account
print("\n### Savings Account Operations ###")
savings.deposit(200)
savings.withdraw(100)
savings.display_details()

# Operations on Checking Account
print("\n### Checking Account Operations ###")
checking.deposit(150)
checking.withdraw(500)  # Insufficient balance example
checking.withdraw(100)
checking.display_details()
```

Question 13:
# Define a class called Triangle with attributes side1, side2, and side3.
# Include methods to check if the triangle is valid, calculate its perimeter, area, and display its sides.
# Create an instance of the Triangle class with side lengths 3, 4, and 5, then check if it is valid, calculate its perimeter and area, and display its sides.

```python
import math

class Triangle:
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    # Check if triangle is valid (Triangle Inequality Theorem)
    def is_valid(self):
        return (self.side1 + self.side2 > self.side3 and
                self.side2 + self.side3 > self.side1 and
                self.side1 + self.side3 > self.side2)

    # Perimeter of triangle
    def perimeter(self):
        return self.side1 + self.side2 + self.side3

    # Area using Heron's Formula
    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

    # Display sides
    def display_sides(self):
        print(f"Sides: {self.side1}, {self.side2}, {self.side3}")


# Create an instance with sides 3, 4, 5
triangle = Triangle(3, 4, 5)

# Display sides
triangle.display_sides()

# Check validity
if triangle.is_valid():
    print("Triangle is valid.")
    print("Perimeter:", triangle.perimeter())
    print("Area:", triangle.area())
else:
    print("Triangle is NOT valid.")
```
