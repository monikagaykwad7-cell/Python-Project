Filtering & Conditional Selection

```python
# 11.	Find all employees who achieved their sales target
df[df['TargetAchieved'] == "Yes"]
```

```python
# 12.	Find employees who belong to the "Electronics" department.
df[df['Department'] == "Electronics"]
```

```python
# 13.	Find employees who belong to the "West" region and achieved their target
df[(df['Region'] == "West") & (df['TargetAchieved'] == "Yes")]
```

```python
# 14.	Filter employees who have a SalesAmount greater than 30,000.
df[df['SalesAmount'] > 30000]
```

```python
# 15.	Find employees who have a commission percentage greater than 5%.
df[df['CommissionPercentage'] > 5]
```

```python
# 16.	Find employees who did not achieve their target.
df[df['TargetAchieved'] == "No"]
```

```python
# 17.	Find employees who belong to either "Groceries" or "Automobile" department.
df[(df['Department'] == "Groceries") | (df['Department'] == "Automobile")]
```

```python
# 18.	Find employees with TotalEarnings greater than 40,000.
df[df['TotalEarnings'] > 40000]
```

```python
# 19.	Find employees who are older than 40 years and achieved their target.
df[(df['Age'] > 40) & (df['TargetAchieved'] == "Yes") ]
```

```python
# 20. Select employees whose names contain "10".
df[df['Name'].str.contains("10")]
```

Aggregation & Grouping

```python
# 21.	Find the average sales amount per department.
df.groupby('Department')['SalesAmount'].mean()
```

```python
# 22.	Count the number of employees in each region.
df.groupby("Region")["EmployeeID"].count()
```

```python
# 24.	Find the lowest SalesAmount in each department
df.groupby("Department")["SalesAmount"].min()
```

```python
# 23.	Find the highest SalesAmount in each department.
df.groupby("Department")["SalesAmount"].max()
```

```python
# 25.	Find the employee with the highest total earnings.
df['SalesAmount'].max()
df[df['SalesAmount'] == df['SalesAmount'].max()]
```

```python
# 27.	Find the average commission percentage per department.
df.groupby("Department")["CommissionPercentage"].mean()
```

```python
# 26.	Count the number of employees who achieved and did not achieve their target
df.groupby("TargetAchieved")['TargetAchieved'].count()
```

```python
# 28.	Find the highest TotalEarnings per region.
df.groupby("Region")['TotalEarnings'].max()
```

```python
# 29.	Find the lowest TotalEarnings per department.
df.groupby("Department")['TotalEarnings'].min()
```

```python
# 30.	Count the number of employees per department.
df.groupby("Department")['EmployeeID'].count()
```
